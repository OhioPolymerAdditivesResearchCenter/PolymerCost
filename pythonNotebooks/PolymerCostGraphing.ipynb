```python
#simple 2d graphing using PolymerCost by OHPARC

#import matplotlib and numpy to handle all the other math and graphics
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from matplotlib.animation import PillowWriter


#these are just three functions from PolymerCost we are using
def flex_clear_dinp(phr):
    duro = 113.8 - 0.6*phr
    modulus = 3559.1 - 34.78*phr
    tensile = 4414.8 - 28.7*phr
    elongation = 248.6 + 1.61*phr
    clashberg = 14.23 - 0.76*phr
    brittle = -4.19 - 0.55*phr
    output = (duro, modulus, tensile, elongation, clashberg, brittle)
    return output

def flex_clear_dotp(phr):
    duro = 113.8 - 0.617*phr
    modulus = 3650.3-35.4*phr
    tensile = 4419.7-26.8*phr
    elongation = 250.9+1.96*phr
    clashberg = 10.8-.77*phr
    brittle = -4.7-.57*phr
    output = (duro, modulus, tensile, elongation, clashberg, brittle)
    return output

def flex_clear_dinch(phr):
    duro = 115.2 - 0.602*phr
    modulus = 3124.3-28.3*phr
    tensile = 4274.7-26.3*phr
    elongation = 335.1+.6*phr
    clashberg = 11.64-.79*phr
    brittle = -6.71-.57*phr
    output = (duro, modulus, tensile, elongation, clashberg, brittle)
    return output

# linear spaced numbers from 20 to 120, 100 of them, for phr's
#this is the range of PHR's of Plasticizer we will be passing to the functions and then graphing
x = np.linspace(20,120,100)
a = np.linspace(20,120,100)
ying = np.linspace(20,120,100)

#here we are running those x, a, and ying linear spaced numbers through the function and returning the first [zeroth in python] term which is the durometer
y = flex_clear_dinch(x)[0]
b = flex_clear_dotp(x)[0]
yang = flex_clear_dinp(ying)[0]

# setting up the plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# plot the function
plt.plot(x,y,'k-', label='DINCH')
plt.plot(a,b,'k--', label='DOTP')
plt.plot(ying,yang,'k:', label="DINP")
#label x, y, and add a legend
plt.xlabel("PHR Plasticizer")
plt.ylabel("A Shore Durometer @ 15s")
leg = ax.legend()

# show the plot
plt.show()
fig.savefig('graphDurometer.png')

#you can find the file you saved in the same folder this python script is
#if you are running this in Jupyter Notebook or Google Colab, check out the left bar and click on the files icon (4th down, looks like a briefcase)
#your file is in there if you want to download it locally
```
